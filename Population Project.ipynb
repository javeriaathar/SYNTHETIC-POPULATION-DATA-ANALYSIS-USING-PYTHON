## IMPORTING LIBRARIES 

```python
import pandas as pd
import numpy as np
import random
```

* pandas (pd): Used for handling and analyzing structured data (like tables/dataframes).
* numpy (np): Used for numerical operations, arrays, and mathematical computations.
* random: Used to generate random values for simulations or data creation.

## STEP 1: CREATE DATASET

```python
np.random.seed(42)

n = 1000

data = []

for i in range(n):
    age = np.random.randint(1, 80)
    gender = random.choice(["Male", "Female"])
    city = random.choice(["Karachi", "Lahore", "Islamabad", "Peshawar"])
    
    data.append([i+1, age, gender, city])

df = pd.DataFrame(data, columns=["ID", "Age", "Gender", "City"])

df.to_csv("population_1000.csv", index=False)

print("Dataset Created Successfully!")
```

* This code creates a synthetic (fake) population dataset of 1000 records and saves it as a CSV file named population_1000.csv.
* It generates:
     * ID ---> Unique number (1 to 1000)
     * Age ---> Random age between 1 and 79
     * Gender ---> Randomly selected (Male/Female)
     * City ---> Randomly selected (Karachi, Lahore, Islamabad, Peshawar)
* Finally, it stores everything in a pandas DataFrame and exports it to a CSV file.

## STEP 5: VISUALIZATION

### 1.  GENDER DISTRIBUTION (BAR CHART)

```python
import matplotlib.pyplot as plt

df["Gender"].value_counts().plot(kind="bar")
plt.title("Gender Distribution in Population")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

* This code creates a bar chart showing the distribution of Gender in your dataset using matplotlib.
* Counts how many Male and Female entries exist.
* Visualizes these counts as a bar chart.
* Add labels for clarity.
* Displays the final graph.

### Insights:
* The dataset shows the distribution of genders (Male vs Female).
* It helps identify whether the dataset is balanced or imbalanced.
* It provides a quick demographic overview of the population.
* This can help in understanding data bias before further analysis or modeling.
##### "This is a basic data visualization to understand gender distribution in your dataset."

### 2.  AGE DISTRIBUTION (HISTOGRAM)

```python
plt.hist(df["Age"], bins=10, edgecolor="black")

plt.title("Age Distribution in Population")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

* This code creates a histogram of Age distribution in your dataset using matplotlib.
* Selects the Age column.
* Divides ages into 10 intervals (groups).
* Makes bars clearer and easier to read.
* The graph shows how often different age ranges appear.

### Insights:
* The graph shows the distribution of different age groups in the dataset.
* It helps identify the most common age range (e.g., children, young adults, adults).
* It reveals the overall pattern of the data (skewed, uniform, or normal-like).
* It provides insight into the population structure of the dataset.
* Age distribution understanding is useful for feature analysis in machine learning.
##### "This is a distribution analysis tool to understand how ages are spread across your dataset."

### 3.  COMBINED ANALYSIS

### Gender + Age Insight

```python
import seaborn as sns

sns.boxplot(x="Gender", y="Age", data=df)

plt.title("Age Distribution by Gender")
plt.show()
```

* This code creates a boxplot showing Age distribution across Gender groups using seaborn
* Compares Age differences between Male and Female.
* Each box shows:
    * Middle line ---> median age
    * Box ---> middle 50% of data (IQR)
    * Whiskers ---> range of typical values
    Dots (if any) ---> outliers (unusual ages)

### Insights
* One gender may have a higher or lower median age compared to the other.
* It shows the spread (variability) of ages within each gender group.
* It helps identify outliers, such as unusually young or old individuals.
* It provides a comparison of age distribution between male and female passengers.
* Useful for understanding demographic differences in the dataset.
